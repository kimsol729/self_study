### 3D Axes (of class Axes3D)

3D Axes (of class Axes3D) are created by passing the projection="3d" keyword argument to Figure.add_subplot:

ax.plot_surface(X,Y,Z) : X,Y는 meshgrid and Z는 X,Y로 이루어진 multivariable ftn

    - rstride : 기본값이 10 인 행 단계 크기
    - cstride : 기본값이 10 인 열 단계 크기
    - cmap : surface의 color map
    - linewidth :
    - antialiased : 앤티앨리어싱 false로 설정함으로써 불투명효과 (곡면 안쪽 모습이 안보임)
    - color : 표면의 색상
cmap : 표면의 컬러 맵
facecolors : 표면의 각 패치에 대한 페이스 컬러
norm : 값을 색상에 매핑하는 Normalize 인스턴스
vmin : 매핑 할 최소값
vmax : 매핑 할 최대 값
shade : face 색상을 음영할지 여부

```python
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np
```

```python
# from mpl_toolkits.mplot3.axes3d import get_test_data


fig = plt.figure(figsize = plt.figaspect(0.5))

# ==============
# First subplot
# ==============
ax = fig.add_subplot(1,2,1,projection = '3d')

X = np.arange(-5,5,0.25)
Y = np.arange(-5,5,0.25)
X,Y = np.meshgrid(X,Y)
R = np.sqrt(X**2+Y**2)
Z = np.cos(R)

surf = ax.plot_surface(X,Y,Z,rstride = 1,cstride = 1, cmap = cm.coolwarm, linewidth = 0,antialiased = False)
ax.set_zlim(-1.01,1.01)
fig.colorbar(surf,shrink = 0.5,aspect = 10)

# ==============
# Second subplot
# ==============
ax = fig.add_subplot(1,2,2,projection = '3d')
surf = ax.plot_wireframe(X,Y,Z,rstride = 2,cstride = 2)
ax.set_zlim(-1.01,1.01)
```

# Line plots

Axes3D.plot(xs,ys,\*args, zdir = 'z',\**kwargs)

```python
ax = plt.figure().add_subplot(projection = '3d')

t = np.linspace(-4*np.pi , 4*np.pi , 100) # 매개변수 t

z = np.linspace(-2,2,100)
r = z**2+1
x = r*np.sin(t) # x = (t^2+1)sint
y = r*np.cos(t) # y = (t^2+1)cost


ax.plot(x,y,z, label = 'parameteric curve')
ax.legend()

plt.show()
```

### Scatter plot

```python
np.random.seed(19680801)

def randrange(n,vmin,vmax):
    """
    Helper fuction to maje an array of random numbers having shape(n, )
    with each number distribution Unif(vmin,vmax).
    """
    return (vmax-vmin)*np.random.rand(n) + vmin

fig = plt.figure()
ax = fig.add_subplot(projection = '3d')

n = 100

for m,zlow,zhigh in [('o',0,10),('^',30,40)]:
    xs = randrange(n,0,100) # x 의 범위 0<x<100
    ys = randrange(n,0,100) # y 의 범위 0<y<100
    zs = randrange(n,zlow,zhigh) # z 의 범위 0<z<10 -> 30<z<40
    ax.scatter(xs, ys, zs,marker = m)
    
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```
